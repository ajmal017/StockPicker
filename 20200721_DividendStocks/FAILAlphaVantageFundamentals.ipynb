```python
def get_CF_reports(ticker):
    allCF = requests.get(f'https://www.alphavantage.co/query?function=CASH_FLOW&symbol={ticker}&apikey={api_key}').json() 
    
    annual = allCF['annualReports']
    dfAnnual = pd.DataFrame.from_dict(annual)
    
    quarterly = allCF['quarterlyReports']
    dfQuarterly = pd.DataFrame.from_dict(quarterly)
    
    return dfAnnual, dfQuarterly
```
